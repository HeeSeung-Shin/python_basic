### 데이터 Merge
- merge()함수
    - sql의 join과 비슷
    - 특정한 column을 기준으로 병합
    - how 파리미터를 통해 병합 방법 지정
        - inner : default. 일치하는 값이 있는 경우 병합
        - left : leftouter join
        - right : right outer join
        - outer : full outer join
    - on 파라미터를 통해 join기준이 되는 column 지정

#### 상하merge

```python
df1 =pd.read_spss("bank1.sav")
df1
```

```python
df2 =pd.read_spss("bank2.sav")
df2
```

```python
#합집합
pd.merge(df1, df2, how="outer") 
```

```python
#교집합
pd.merge(df1, df2, how= "inner")
```

#### 좌우merge

```python
df3 = pd.read_spss("bank3.sav")
df3
```

```python
df4= pd.read_spss("bank4.sav")
df4
```

```python
# on=id id기준으로 옆으로 합침
pd.merge(df3, df4, on = "id")
```

### 데이터 분석하기
- #### 기초통계 분석하기 

- 척도
    - 척도,자료의 종류가 무엇인가에 따라 분석방법이 결정됨
    - 따라서 척도와 자료의 종류를 이해하는 것은 분석에 앞서 매우 중요함
    - 척도(scale)란 사물이나 사람의 특성(속성)을 하나의 기준으로 숫자를 부여한 것. 측정의 단위 도구
    - 범주형(이산형)변수인 경우에는 각 변수에 대한 빈도표와 막대그래프,원그래프를 통해 빈도분석을 시행
    - 연속형인 경우에는 일반적으로 평균, 중앙값, 최빈값, 분산, 히스토그램등을 통해 변수의 분포를 확이함 

#### 척도의 종류
- 범주형 자료(분석법:빈도 분석)
    - 명목척도:측정대상을 단지 분료할 목적으로 숫자를 부여한 척도 ex)성별,등번호
    - 서열척도: 측정대상간에 순서/서열적 의미가 있는 척도 ex)학력,선호도
- 연속형 자료(분석법: 기술통계)
    - 등간척도:측정대상간에 순서 뿐만 아니라 숫자의 간격이 동일하여 양적인 정도를 알 수 잇는 척도 ex)온도,점수,만족도
    - 비율척도:등간척도와 유사하나 0의 의미가 '실제로 없다'고 말할 수 있는 척도 ex)무게 ,길이,소득


#### 변수의 종류
- 변수: 변하는 모든 수를 말함. ex)사람의 키, 체중
    - 독립변수:다른 변수에 영향을 주는 변수
    - 종속변수: 독립변수에 의하여 변화되는 변수

#### 자료의 종류에 따른 분석법
- 독:범주/종:범주-> 교차분석(카이제곱 분석)
- 독:범주/종:연속-> t/test/ANOVA(분산분석)
- 독:연속/종:범주-> 로지스틱 회귀분석 판별분석
- 독:연속/종:연속-> 상관분석, 회귀분석

## 상자그래프 점보
- max(최대값): 사분위 범수에 1.5배 떨어진값
- Q3:중앙값과 최대값의 중간값
- Q2:중앙값
- Q1:중앙값과 최소값의 중간값
- min(최소값):사분위 범수에 1.5배 떨어진 값
- IQR(산분위 범수):Q3-Q1
- outlier:사분위범수 1.5배 값에 들어오지 않는 값

```python
Q1=bank['age'].quantile(q=0.25)
Q3=bank['age'].quantile(q=0.75)
IQR = Q3-Q1
print(Q1, Q3, IQR)
```

```python
bank["age"].describe()
```

```python
minimum = Q1-1.5*IQR
maximum = Q3+1.5*IQR
print(minimum,maximum)
#19 minimum 값 74~87까지 이상치
```
