```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
epl = pd.read_csv("epl.csv")
```

```python
epl=epl.fillna(0)
```

```python
epl
```

```python
epl["Assist per match"]=(epl["Assists"])/(epl["Appearances"])
```

```python
epl["Shots on target per match"]=(epl["Shots on target"])/(epl["Appearances"])
```

```python
epl["Yellow cards per match"]=(epl["Yellow cards"])/(epl["Appearances"])
```

```python
epl["Red cards per match"]=(epl["Red cards"])/(epl["Appearances"])
```

```python
epl["Tackles per match"]=(epl["Tackles"])/(epl["Appearances"])
```

```python
epl["Interceptions per match"]=(epl["Interceptions"])/(epl["Appearances"])
```

```python
epl["Blocked shots per match"]=(epl["Blocked shots"])/(epl["Appearances"])
```

```python
epl["Clearances per match"]=(epl["Clearances"])/(epl["Appearances"])
```

```python
epl["Duels won per match"]=(epl["Duels won"])/(epl["Appearances"])
```

```python
epl["Errors leading to goal per match"]=(epl["Errors leading to goal"])/(epl["Appearances"])
```

```python
epl_Forward= epl[epl["Position"]=="Forward"]
```

```python
epl_Midfielder= epl[epl["Position"]=="Midfielder"]
```

```python
epl["Yellow cards per match"]=(epl["Yellow cards"])/(epl["Appearances"])
```

```python
epl_Defender= epl[epl["Position"]=="Defender"]
```

```python
epl_Goalkeeper= epl[epl["Position"]=="Goalkeeper"]
```

```python
epl_Forward=epl_Forward[epl_Forward.Appearances>100].reset_index()
epl_Forward=epl_Forward[epl_Forward.Appearances>100].drop(["index"], axis=1) 
```

```python
epl_Midfielder=epl_Midfielder[epl_Midfielder.Appearances>100].reset_index()
epl_Midfielder=epl_Midfielder[epl_Midfielder.Appearances>100].drop(["index"], axis=1) 
```

```python
epl_Defender=epl_Defender[epl_Defender.Appearances>100].reset_index()
epl_Defender=epl_Defender[epl_Defender.Appearances>100].drop(["index"], axis=1) 
```

```python
epl_Goalkeeper=epl_Goalkeeper[epl_Goalkeeper.Appearances>100].reset_index()
epl_Goalkeeper=epl_Goalkeeper[epl_Goalkeeper.Appearances>100].drop(["index"], axis=1) 
```

```python
epl_Forward.columns
```

```python
epl_Forward_best=epl_Forward[["Name","Goals per match","Assist per match","Shots on target per match","Yellow cards per match","Red cards per match"]]
```

```python
epl_Forward_best["Total Score"]=epl_Forward_best["Goals per match"]*1+epl_Forward_best["Assist per match"]*1+epl_Forward_best["Shots on target per match"]*0.5-epl_Forward_best["Yellow cards per match"]*5-epl_Forward_best["Red cards per match"]*10
```

```python
epl_Forward_best.sort_values(by='Total Score',ascending=False)
```
